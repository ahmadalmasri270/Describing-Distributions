```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
plt.style.use(['seaborn-muted','fivethirtyeight'])
sns.set_context('talk')
```

### Exploring Distributions

```python
df = pd.read_csv('Data/Medical Dataset.csv')
df.head()
df.info()
```

```python
stats= df['VitD_levels'].agg(['mean','std','min','max'])
stats
```

```python
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
sns.displot(x= df.VitD_levels,kde=True , discrete=True, stat='density')
plt.axvline(mean, c='r', ls='-', lw=2.5)
plt.axvline(median, c='g', ls='--', lw=2.5)
plt.axvline(plus_one_std, c='k', ls='--', lw=2.5)
plt.axvline(minus_one_std, c='k', ls='--', lw=2.5)
plt.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
```

```python
skewValue = df['VitD_levels'].skew()
print("Skew:")
print(skewValue)
```

```python
from scipy import stats
kurt = stats.kurtosis(df['VitD_levels'], fisher = False)
kurt
```
